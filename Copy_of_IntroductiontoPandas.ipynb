### Data exploration


```python
import pandas as pd

data = {
    "Name": ["Hala", "Ameer", "Lina", "Ali", "Sara"],
    "Age": [22, 25, 30, 28, 35],
    "City": ["Nablus", "Ramallah", "Hebron", "Jenin", "Tulkarm"]
}

df = pd.DataFrame(data)
display(df)
df.head()        # أول 5 صفوف
df.tail()        # آخر 5 صفوف
df.info()        # معلومات عن الأعمدة والأنواع
df.describe()    # إحصائيات (متوسط، انحراف معياري...)
df.shape         # (عدد الصفوف, عدد الأعمدة)
df.columns       # أسماء الأعمدة
df.dtypes        # نوع بيانات كل عمود

```

###Data cleaning

```python
# القيم الفارغة
df.dropna()                 # حذف الصفوف اللي فيها NaN
df.fillna(0)                # استبدال NaN بالقيمة 0

# التكرارات
df.drop_duplicates()

# تعديل الأعمدة
df.rename(columns={"Name" : "FullName"}, inplace=True)
df["Age"] = df["Age"].astype(float)   # تغيير نوع البيانات
```

###Simple analysis

```python
df.sort_values("Age", ascending=False)   # ترتيب بالعمر

df["City"].value_counts()                # تكرار القيم

df.groupby("City")["Age"].mean()         # متوسط العمر لكل مدينة

df["Age"].mean()    # متوسط
df["Age"].sum()     # مجموع
df["Age"].max()     # أكبر قيمة
df["Age"].min()     # أصغر قيمة
```

###Select and separate data


```python
# عمود واحد أو أكثر
df["Name"]
df[["Name", "City"]]

# حسب الفهرس
df.iloc[0]       # الصف الأول
df.iloc[0:3]     # أول 3 صفوف

# حسب الاسم
df.loc[0, "Name"]   # قيمة محددة
df.loc[0:2, ["Name","Age"]]

# شروط
df[df["Age"] > 25]
df[(df["Age"] > 20) & (df["City"] == "Nablus")]
```
